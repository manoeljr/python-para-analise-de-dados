```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go

import warnings

warnings.filterwarnings('ignore')
```

```python
base_dados = pd.read_excel('/content/drive/MyDrive/Estudos/Udemy/Python para Análise de dados/Dados/Vase_004 - Magalu - Sem Resolução.xlsx')
base_dados.head()
```

```python
# Colocando a coluna data como index para fazer analise com series temporais
dados = base_dados.set_index('Data')
dados.head()
```

```python
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(16, 5))
plt.title('Analise das ações da MagaLu - Fechamento', fontsize=15, loc='left')
plt.plot(dados.index, dados.Fechamento)

plt.xlabel('Período da Cotação')
plt.ylabel('Valor da Ação (R$)');
```

```python
dados.tail()
```

```python
# Media movel de 5 dias
media_movel_5_dias = dados.Fechamento.rolling(5).mean()

# Media movel tendencia, 30 dias
media_movel_30_dias = dados.Fechamento.rolling(30).mean()

# Plotando as media
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(16, 5))
plt.title('Analise das ações da MagaLu - Fechamento', fontsize=15, loc='left')

plt.plot(dados.index, dados.Fechamento)
plt.plot(dados.index, media_movel_5_dias)
plt.plot(dados.index, media_movel_30_dias, linestyle='dashed')

plt.xlabel('Período da Cotação')
plt.ylabel('Valor da Ação (R$)');
```

```python
sns.boxplot(data=dados, x='Fechamento');
```

```python
# Boxplot Mensal
base_dados['Mes'] = base_dados['Data'].dt.month


plt.figure(figsize=(16, 5))
sns.boxplot(data=base_dados, x='Mes', y='Fechamento');
```

```python
base_dados.groupby(['Mes']).describe()['Fechamento']
```

```python
from matplotlib.axis import XAxis
grafico = go.Figure(
    data=[
          go.Candlestick(
              x = dados.index,
              open = dados.Abertura,
              high = dados.Maior,
              low = dados.Menor,
              close = dados.Fechamento,
          )
    ]
)

grafico.update_layout(xaxis_rangeslider_visible=False)

grafico.show()
```
